```python
class TrafficDevice:
    """Parent base class for all traffic devices."""
    def activate(self):
        print("Traffic device activated.")


class TrafficLight(TrafficDevice):
    """Child class representing a traffic light."""
    def activate(self):
        print("Traffic Light: Changing signal state (Red -> Green).")


class SpeedCamera(TrafficDevice):
    """Child class representing a speed camera."""
    def activate(self):
        print("Speed Camera: Monitoring vehicle speeds and capturing licenses.")


class PedestrianSignal(TrafficDevice):
    """Child class representing a pedestrian signal."""
    def activate(self):
        print("Pedestrian Signal: Displaying 'WALK' signal for pedestrians.")


class EmergencySiren(TrafficDevice):
    """
    New class added dynamically. Inherits from TrafficDevice to seamlessly
    fit into the existing activation loop without code modifications.
    """
    def activate(self):
        print("Emergency Siren: Sounding alarm and flashing warning lights!")



# Execution / Activation Demonstration

if __name__ == "__main__":

    devices = [
        TrafficLight(),
        SpeedCamera(),
        PedestrianSignal(),
        EmergencySiren()
    ]

    print("--- Activating Smart Traffic System Devices ---")


    for device in devices:
        device.activate()
```
